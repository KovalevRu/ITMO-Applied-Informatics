В этом разделе необходимо обработать значения в трех столбцах датасета: `location`, `salary` и `job_type`. Цель обработки – стандартизировать данные в каждом из столбцов так, чтобы их можно было анализировать.

### Задание 1

В задании 1 необходимо провести обработку столбца `location`.

#### Задание 1.1

В столбце `location` представлена информация о местоположении организации, в которой открыта вакансия. Каждое значение столбца `location` содержит информацию о городе, коде штата и почтовом индексе.

Однако не в каждой строке есть все три характеристики местоположения, а также могут встречаться разные комбинации из них, записанные в виде строки в произвольном порядке и с разными разделителями.

Например, может быть указан только код штата, только название города или только индекс. Могут быть указаны любые две характеристики из трех.

**Формулировка задания:**
1. Определите, какие форматы встречаются.
2. Напишите одно или несколько регулярных выражений для поиска выявленных форматов.
3. Визуализируйте количество найденных форматов на столбчатой или круговой диаграмме. Также покажите на ней количество строк, в которых формат определить не удалось. Какие форматы встречаются чаще всего?

Ознакомиться с теоретическим материалом по работе с регулярными выражениями можно [по ссылке](https://colab.research.google.com/drive/1eFHdmt9OROLEDJRNc2wNmz0gpE1O3sKU?usp=sharing).

````python
# Смотрим на примеры location
print(df['location'].value_counts().head(20))
print()

# Форматы:
# 1. City, ST XXXXX — город, код штата, индекс
# 2. City, ST — город, код штата
# 3. City — только город
# 4. ST XXXXX — код штата и индекс
# 5. Что-то другое (длинный текст из job_description и т.д.)

patterns = {
    'City, ST XXXXX': r'^[^,]+,\s*[A-Z]{2}\s+\d{5}$',
    'City, ST': r'^[^,]+,\s*[A-Z]{2}$',
    'City only': r'^[A-Za-z\s\-\.]+$',
    'ST XXXXX': r'^[A-Z]{2}\s+\d{5}$',
    'ZIP only': r'^\d{5}$',
}

location = df['location'].fillna('')

counts = {}
matched = pd.Series([False] * len(df))
for name, pat in patterns.items():
    mask = location.str.match(pat, na=False)
    # не считаем уже найденные
    new_mask = mask & ~matched
    counts[name] = new_mask.sum()
    matched |= new_mask

counts['Не определён'] = (~matched).sum()

print(pd.Series(counts))

# График
pd.Series(counts).plot(kind='bar', figsize=(10, 4), color='steelblue')
plt.title('Форматы значений в столбце location')
plt.ylabel('Количество')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
````

#### Задание 1.2

В большей части строк удалось извлечь данные из ячейки столбца `location`, но остались и такие, где это не получилось. Вынесите в отдельный датасет записи, в которых не удалось определить формат записи информации о местоположении.

Над оставшимися строками проведите следующие преобразования:
1. Данные о местоположении разделите на 3 столбца: город, код штата, почтовый индекс.
2. Так как значения в столбце `почтовый индекс` являются пятизначными числами, преобразуйте их в целочисленный формат.

````python
# Разделяем по паттернам location на 3 столбца

location = df['location'].fillna('')

# Паттерн: City, ST XXXXX или City, ST или ST XXXXX
full_pattern = r'^([^,]+),\s*([A-Z]{2})\s+(\d{5})$'
city_state_pattern = r'^([^,]+),\s*([A-Z]{2})$'
state_zip_pattern = r'^([A-Z]{2})\s+(\d{5})$'

def parse_location(loc):
    loc = str(loc).strip()
    m = re.match(full_pattern, loc)
    if m:
        return m.group(1).strip(), m.group(2).strip(), m.group(3)
    m = re.match(city_state_pattern, loc)
    if m:
        return m.group(1).strip(), m.group(2).strip(), None
    m = re.match(state_zip_pattern, loc)
    if m:
        return None, m.group(1), m.group(2)
    return None, None, None

parsed = location.apply(parse_location)
df[['city', 'state', 'zip']] = pd.DataFrame(parsed.tolist(), index=df.index)

# Определяем, какие строки не распарсились
not_parsed_mask = df['city'].isna() & df['state'].isna() & df['zip'].isna()
df_not_parsed = df[not_parsed_mask].copy()
df_main = df[~not_parsed_mask].copy()

print(f'Основная выборка: {len(df_main)}')
print(f'Не удалось определить: {len(df_not_parsed)}')

# Преобразуем zip в int там где есть
df_main['zip'] = pd.to_numeric(df_main['zip'], errors='coerce').astype('Int64')

df_main[['location', 'city', 'state', 'zip']].head(5)
````

Рассмотрите отложенную часть выборки (ту, где не удалось данные из столбца `location` разделить на три поля) и сделайте выводы о возможности извлечь из данных город, код штата и почтовый индекс. Извлеките все, что возможно. Что невозможно – оставьте пустым.

````python
# Работаем с df_not_parsed — строки, которые не подошли под стандартные паттерны
print(f'Строк в отложенной части: {len(df_not_parsed)}')
print(df_not_parsed['location'].head(10).tolist())

# Пытаемся достать хоть что-то
def parse_edge_case(loc):
    loc = str(loc).strip()
    city, state, zip_code = None, None, None
    
    # ищем zip
    m_zip = re.search(r'\b(\d{5})\b', loc)
    if m_zip:
        zip_code = m_zip.group(1)
    
    # ищем код штата
    m_state = re.search(r'\b([A-Z]{2})\b', loc)
    if m_state:
        state = m_state.group(1)
    
    return city, state, zip_code

parsed2 = df_not_parsed['location'].apply(parse_edge_case)
df_not_parsed[['city', 'state', 'zip']] = pd.DataFrame(parsed2.tolist(), index=df_not_parsed.index)
df_not_parsed['zip'] = pd.to_numeric(df_not_parsed['zip'], errors='coerce').astype('Int64')

print(df_not_parsed[['location', 'city', 'state', 'zip']].head(5))
print(f'\\nГород удалось извлечь: {df_not_parsed["city"].notna().sum()}')
print(f'Штат удалось извлечь: {df_not_parsed["state"].notna().sum()}')
print(f'Индекс удалось извлечь: {df_not_parsed["zip"].notna().sum()}')
````

### Задание 2

В задании 2 необходимо провести обработку столбца `salary`.

#### Задание 2.1

В столбце `salary`, как и в `location`, данные не стандатизированы. В нем в основном можно встретить записи в следующих форматах:

* 10.00 - 20.00 &dollar; &sol;hour
* 10,000.00 - 20,000.00 &dollar; &sol;year
* &dollar;10.00 &sol;hour
* &dollar;10,000.00+ &sol;year
* Up to &dollar;20,000.00

Основные проблемы, которые стоит учесть при обработке форматов:
- В каждом из форматов может быть или не быть знака доллара.
- Может отсутствовать информация о периоде, за который указывается зарплата (`/hour` или `/year` в примерах).
- Зарплата может быть указана за месяц (`/month`) и за неделю (`/week`).
- В начале, конце и середине некоторых строк есть лишние пробелы, которые не стоит учитывать.

**Задание:**
1. Определите весь перечень встречающихся форматов и напишите одно или несколько регулярных выражений для их поиска.
2. Покажите на столбчатой диаграмме количество записей, содержащих каждый из форматов. Сделайте выводы.

````python
salary = df['salary'].fillna('')

patterns_salary = {
    'MIN - MAX $ /period': r'\d[\d,\.]*\s*-\s*\d[\d,\.]*\s*\$\s*/\w+',
    '$ MIN - MAX /period': r'\$\s*\d[\d,\.]*\s*-\s*\$?\s*\d[\d,\.]*\s*/\w+',
    '$ VALUE /period': r'\$[\d,\.]+\+?\s*/\w+',
    'VALUE $ /period': r'\d[\d,\.]*\+?\s*\$\s*/\w+',
    'Up to $ VALUE': r'[Uu]p\s+to\s+\$[\d,\.]+',
    'Без числового значения': r'^[^\d]+$',
}

counts_s = {}
matched_s = pd.Series([False] * len(df))
for name, pat in patterns_salary.items():
    mask = salary.str.contains(pat, na=False, regex=True)
    new_mask = mask & ~matched_s
    counts_s[name] = new_mask.sum()
    matched_s |= new_mask

# пустые
empty_mask = salary == ''
counts_s['Пустые'] = (empty_mask & ~matched_s).sum()
matched_s |= empty_mask

counts_s['Другое'] = (~matched_s).sum()

print(pd.Series(counts_s))

pd.Series(counts_s).plot(kind='bar', figsize=(10, 4), color='coral')
plt.title('Форматы значений в столбце salary')
plt.ylabel('Количество')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
````

#### Задание 2.2

1. Разделите столбец `salary` зарплаты три: `минимальная граница зарплаты`, `максимальная граница зарплаты`, `период`. Если значение суммы указано только одно, считайте, что это минимальная зарплата. Значение суммы в обоих столбцах преобразуйте в формат `float`.
2. Покажите на столбчатой диаграмме с группировкой среднюю и медианную минимальную зарплату в каждом периоде. Сделайте выводы.

````python
salary = df['salary'].fillna('')

def parse_salary(s):
    s = str(s).strip()
    if not s:
        return np.nan, np.nan, None
    
    # ищем период
    period_match = re.search(r'/\s*(hour|year|month|week)', s, re.IGNORECASE)
    period = period_match.group(1).lower() if period_match else None
    
    # ищем числа
    nums = re.findall(r'[\d,]+\.?\d*', s.replace(',', ''))
    nums = [float(n.replace(',', '')) for n in nums if n]
    
    if len(nums) == 0:
        return np.nan, np.nan, period
    elif len(nums) == 1:
        return nums[0], np.nan, period
    else:
        return nums[0], nums[1], period

parsed_salary = df['salary'].apply(parse_salary)
df[['salary_min', 'salary_max', 'salary_period']] = pd.DataFrame(parsed_salary.tolist(), index=df.index)

# График средней и медианной зарплаты по периодам
period_stats = df[df['salary_period'].notna()].groupby('salary_period')['salary_min'].agg(['mean', 'median'])
period_stats.columns = ['Средняя', 'Медиана']
print(period_stats)

period_stats.plot(kind='bar', figsize=(8, 4))
plt.title('Средняя и медианная минимальная зарплата по периодам')
plt.ylabel('Зарплата')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
````

#### Задание 2.3

Как было отмечено в задании 2.1, период, за который указывается зарплата, может отсутствовать. Восполните этот пробел в данных следующим образом:

1. Определите диапазон изменения (разброс) значений денежной суммы за каждый из встречающихся в данных периодов. Постройте гистограмму для каждого периода.
2. Определите, к среднему значению за какой период ближе всего сумма, у которой период не указан. Укажите для найденный период. В спорных случаях, где сумма одинаково близка к нескольким диапазонам, не выбирайте ничего.

````python
# Записи без периода
no_period = df[df['salary_period'].isna() & df['salary_min'].notna()].copy()
print(f'Записей без периода: {len(no_period)}')

# Диапазоны значений по периодам
for p in ['hour', 'year', 'month', 'week']:
    vals = df[df['salary_period'] == p]['salary_min'].dropna()
    if len(vals):
        print(f'{p}: min={vals.min():.0f}, max={vals.max():.0f}, mean={vals.mean():.0f}')
        plt.figure(figsize=(6, 3))
        vals[vals < vals.quantile(0.99)].hist(bins=30)
        plt.title(f'Гистограмма salary_min для периода: {p}')
        plt.xlabel('Зарплата')
        plt.ylabel('Кол-во')
        plt.tight_layout()
        plt.show()

# Для каждой записи без периода — определяем ближайший период по среднему
period_means = df[df['salary_period'].notna()].groupby('salary_period')['salary_min'].mean()
print('\nСредние по периодам:')
print(period_means)

def assign_period(val):
    if pd.isna(val):
        return None
    diffs = abs(period_means - val)
    min_diff = diffs.min()
    # если два периода одинаково близко — не выбираем
    if (diffs == min_diff).sum() > 1:
        return None
    return diffs.idxmin()

no_period['predicted_period'] = no_period['salary_min'].apply(assign_period)
df.loc[no_period.index, 'salary_period'] = no_period['predicted_period']
print(f'\nЗаполнено периодов: {no_period["predicted_period"].notna().sum()}')
````

````python
# Смотрим уникальные значения job_type
print(df['job_type'].value_counts().head(30))
print(f'\nВсего уникальных значений: {df["job_type"].nunique()}')
````

Вы заметили, что многие значения фактически одинаковые, но немного по-разному отформатированы? Приведем их к единому формату.

````python
# Стандартизируем job_type: убираем лишние пробелы и запятые
df['job_type_clean'] = df['job_type'].fillna('')
df['job_type_clean'] = df['job_type_clean'].str.strip()
df['job_type_clean'] = df['job_type_clean'].str.replace(r'\s*,\s*', ', ', regex=True)
df['job_type_clean'] = df['job_type_clean'].str.replace(r'\s*/\s*', '/', regex=True)
df['job_type_clean'] = df['job_type_clean'].str.replace(r'\s+', ' ', regex=True)

# Убираем дубль в начале (Job Type Full Time Employee -> Full Time Employee)
df['job_type_clean'] = df['job_type_clean'].str.replace(r'^Job Type\s+', '', regex=True)

print(df['job_type_clean'].value_counts().head(20))
````

Столбец `job_type` состоит из двух частей: "полнота" занятости (`Full Time`, `Part Time`, `Per Diem`) и "тип" занятости (`Employee`, `Intern`). Кроме того, встречаются и неполные записи.

Задание:
1. Написать одно или несколько регулярных выражений для разделения столбца `job_type` на два: `полнота занятости`, `тип занятости`.
2. Если данные о полноте отсутствуют, внесите значение `Full Time`.
Если отсутствуют данные о типе – внесите `Employee`.
3. Постройте столбчатую диаграмму со стековой группировкой для полноты и типа занятости. Сделайте выводы.

````python
def parse_job_type(jt):
    jt = str(jt).strip()
    
    # Полнота
    fullness = None
    for f in ['Full Time', 'Part Time', 'Per Diem']:
        if f.lower() in jt.lower():
            fullness = f
            break
    
    # Тип
    job_kind = None
    for k in ['Employee', 'Intern', 'Temporary/Contract/Project', 'Seasonal']:
        if k.lower() in jt.lower():
            job_kind = k
            break
    
    if fullness is None:
        fullness = 'Full Time'
    if job_kind is None:
        job_kind = 'Employee'
    
    return fullness, job_kind

parsed_jt = df['job_type_clean'].apply(parse_job_type)
df[['job_fullness', 'job_kind']] = pd.DataFrame(parsed_jt.tolist(), index=df.index)

print(df[['job_fullness', 'job_kind']].value_counts().head(15))

# Стековая диаграмма
pivot = df.groupby(['job_fullness', 'job_kind']).size().unstack(fill_value=0)
pivot.plot(kind='bar', stacked=True, figsize=(9, 5))
plt.title('Полнота и тип занятости')
plt.ylabel('Количество вакансий')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
# Вывод: большинство вакансий — Full Time Employee
````
