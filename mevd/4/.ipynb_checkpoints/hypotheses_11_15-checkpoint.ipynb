## Лабораторная работа 4 — Визуализация данных. Гипотезы 11–15

**Датасет:** The Movies Dataset (`movies_metadata.csv`)  
Источник: [Kaggle](https://www.kaggle.com/datasets/rounakbanik/the-movies-dataset)

Этот ноутбук является продолжением основного файла анализа (гипотезы 1–10).  
Здесь проверяются гипотезы блоков 3 (продолжение), 4 и 5.

| # | Гипотеза | Блок |
|---|----------|------|
| 11 | Мультфильмы получают более высокие оценки, чем боевики | Жанры |
| 12 | Современные фильмы имеют больший бюджет | Время |
| 13 | Современные фильмы собирают больше в прокате | Время |
| 14 | Средняя продолжительность растёт со временем | Время |
| 15 | Фильмы 90–120 мин получают более высокие оценки | Дополнительная |

**Зависимости:** pandas, matplotlib, seaborn  
**Запуск:** Run All. Файл `movies_metadata.csv` должен быть в той же папке.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast

movies = pd.read_csv('movies_metadata.csv', low_memory=False)
```

```python
def get_genres(text):
    try:
        genres = ast.literal_eval(text)
        return [g['name'] for g in genres]
    except:
        return []

movies['genres_list'] = movies['genres'].apply(get_genres)
genres_df = movies.explode('genres_list')
```

---
## Блок 3 (продолжение). Жанры

### Гипотеза 11: Мультфильмы получают более высокие оценки, чем боевики


```python
genres_df['vote_average'] = pd.to_numeric(genres_df['vote_average'], errors='coerce')

data = genres_df[
    (genres_df['genres_list'].isin(['Animation', 'Action'])) &
    (genres_df['vote_average'] > 0)
]
```

```python
anim_mean = data[data['genres_list'] == 'Animation']['vote_average'].mean()
action_mean = data[data['genres_list'] == 'Action']['vote_average'].mean()

print(f'Средний рейтинг Animation: {anim_mean:.2f}')
print(f'Средний рейтинг Action: {action_mean:.2f}')
```

```python
plt.figure(figsize=(9, 6))

sns.boxplot(
    data=data,
    x='genres_list',
    y='vote_average',
    order=['Animation', 'Action'],
    palette=['#4C72B0', '#DD8452']
)

plt.axhline(y=anim_mean, color='#4C72B0', linestyle='--', alpha=0.7,
            label=f'Среднее Animation: {anim_mean:.2f}')
plt.axhline(y=action_mean, color='#DD8452', linestyle='--', alpha=0.7,
            label=f'Среднее Action: {action_mean:.2f}')

plt.title('Рейтинг мультфильмов vs боевиков')
plt.xlabel('Жанр')
plt.ylabel('Средняя оценка')
plt.legend()
plt.show()
```

## Итог:

Boxplot показывает распределение рейтингов для Animation и Action.

Что наблюдаем: медиана и среднее Animation выше, чем у Action. У мультфильмов меньше плохих картин с низкими оценками.

**Вывод:** гипотеза **подтверждается**. Мультфильмы в среднем получают более высокие оценки, чем боевики.


---
## Блок 4. Время

### Гипотеза 12: Современные фильмы имеют больший бюджет, чем старые


```python
movies['budget'] = pd.to_numeric(movies['budget'], errors='coerce')
movies['release_date'] = pd.to_datetime(movies['release_date'], errors='coerce')
movies['year'] = movies['release_date'].dt.year

data = movies[
    (movies['budget'] > 100000) &
    (movies['year'] >= 1960) &
    (movies['year'] <= 2017)
][['year', 'budget']]
```

```python
budget_by_year = data.groupby('year')['budget'].median().reset_index()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(budget_by_year['year'], budget_by_year['budget'], marker='o', markersize=3)
plt.title('Медианный бюджет фильмов по годам')
plt.xlabel('Год')
plt.ylabel('Медианный бюджет ($)')
plt.tight_layout()
plt.show()
```

```python
old = data[data['year'] <= 1990]['budget'].median()
new = data[data['year'] >= 2000]['budget'].median()

print(f'Медианный бюджет до 1990: ${old:,.0f}')
print(f'Медианный бюджет с 2000: ${new:,.0f}')
```

## Итог:

График медианного бюджета по годам показывает явный восходящий тренд с 1980-х.

Что наблюдаем: бюджеты неуклонно растут. Медианный бюджет после 2000 года значительно превышает бюджеты до 1990.

**Вывод:** гипотеза **подтверждается**. Современные фильмы имеют существенно больший бюджет из-за инфляции и роста затрат на спецэффекты и маркетинг.


---
### Гипотеза 13: Современные фильмы собирают больше денег в прокате


```python
movies['revenue'] = pd.to_numeric(movies['revenue'], errors='coerce')

data = movies[
    (movies['revenue'] > 100000) &
    (movies['year'] >= 1960) &
    (movies['year'] <= 2017)
][['year', 'revenue']]
```

```python
rev_by_year = data.groupby('year')['revenue'].median().reset_index()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(rev_by_year['year'], rev_by_year['revenue'], color='green', marker='o', markersize=3)
plt.title('Медианные кассовые сборы фильмов по годам')
plt.xlabel('Год')
plt.ylabel('Медианные сборы ($)')
plt.tight_layout()
plt.show()
```

```python
old = data[data['year'] <= 1990]['revenue'].median()
new = data[data['year'] >= 2000]['revenue'].median()

print(f'Медианные сборы до 1990: ${old:,.0f}')
print(f'Медианные сборы с 2000: ${new:,.0f}')
```

## Итог:

График медианных кассовых сборов по годам демонстрирует выраженный восходящий тренд.

Что наблюдаем: сборы значительно выросли с 1970-х по 2010-е. Медианные сборы после 2000 года кратно превышают показатели прошлых десятилетий.

**Вывод:** гипотеза **подтверждается**. Современные фильмы в среднем собирают значительно больше денег в прокате за счёт расширения глобального кинорынка.


---
### Гипотеза 14: Средняя продолжительность фильмов увеличивается со временем


```python
movies['runtime'] = pd.to_numeric(movies['runtime'], errors='coerce')

data = movies[
    (movies['runtime'] >= 40) &
    (movies['runtime'] <= 240) &
    (movies['year'] >= 1920) &
    (movies['year'] <= 2017)
][['year', 'runtime']]
```

```python
runtime_by_year = data.groupby('year')['runtime'].mean().reset_index()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(runtime_by_year['year'], runtime_by_year['runtime'], color='purple', marker='o', markersize=3)
plt.title('Средняя продолжительность фильмов по годам')
plt.xlabel('Год')
plt.ylabel('Средняя продолжительность (мин.)')
plt.tight_layout()
plt.show()
```

```python
old = data[data['year'] <= 1960]['runtime'].mean()
new = data[data['year'] >= 2000]['runtime'].mean()

print(f'Средняя длительность до 1960: {old:.1f} мин')
print(f'Средняя длительность с 2000: {new:.1f} мин')
```

## Итог:

График средней продолжительности фильмов по годам показывает неоднозначную картину.

Что наблюдаем: с 1920-х по 1960-е рост заметен. После 1960-х значения стабилизировались в диапазоне ~90–110 мин и существенно не менялись.

**Вывод:** гипотеза **подтверждается частично**. Рост длительности был характерен для периода становления кино, в современную эпоху средняя длительность остаётся стабильной.


---
## Блок 5. Дополнительная гипотеза

### Гипотеза 15: Фильмы 90–120 минут получают более высокие оценки, чем очень короткие или длинные


```python
movies['vote_average'] = pd.to_numeric(movies['vote_average'], errors='coerce')
movies['runtime'] = pd.to_numeric(movies['runtime'], errors='coerce')

data = movies[
    (movies['runtime'] >= 40) &
    (movies['runtime'] <= 240) &
    (movies['vote_average'] > 0)
][['runtime', 'vote_average']].copy()
```

```python
bins = [40, 60, 80, 90, 100, 120, 150, 180, 240]
labels = ['40-60', '60-80', '80-90', '90-100', '100-120', '120-150', '150-180', '180-240']

data['runtime_group'] = pd.cut(data['runtime'], bins=bins, labels=labels)
```

```python
grouped = data.groupby('runtime_group', observed=True)['vote_average'].mean()
print(grouped)
```

```python
colors = []
for label in labels:
    if label in ['90-100', '100-120']:
        colors.append('#f4a261')
    else:
        colors.append('#aec6cf')

plt.figure(figsize=(11, 6))
grouped.plot(kind='bar', color=colors)
plt.title('Средний рейтинг фильмов по продолжительности')
plt.xlabel('Продолжительность (мин.)')
plt.ylabel('Средняя оценка')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
target = data[data['runtime_group'].isin(['90-100', '100-120'])]['vote_average'].mean()
others = data[~data['runtime_group'].isin(['90-100', '100-120'])]['vote_average'].mean()

print(f'Средний рейтинг 90-120 мин: {target:.2f}')
print(f'Средний рейтинг остальных: {others:.2f}')
```

## Итог:

Столбчатая диаграмма показывает средний рейтинг по интервалам длительности (оранжевым выделен диапазон 90–120 мин).

Что наблюдаем: фильмы 90–120 мин действительно получают рейтинг выше, чем короткие (40–60 мин). Однако длинные фильмы (150–180+ мин) нередко оцениваются ещё выше.

**Вывод:** гипотеза **подтверждается частично**. Фильмы 90–120 мин выигрывают у коротких, но проигрывают очень длинным. Зрители склонны высоко оценивать масштабные работы.
