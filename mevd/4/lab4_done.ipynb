## Блок 1. Финансовый успех фильма

### Гипотеза 1: Чем больше бюджет фильма, тем выше его кассовые сборы.

```python
# Подготовка данных
movies = pd.read_csv("movies_metadata.csv", low_memory=False)

movies["budget"] = pd.to_numeric(movies["budget"], errors="coerce")
movies["revenue"] = pd.to_numeric(movies["revenue"], errors="coerce")

data = movies[
    (movies["budget"] >= 10000) &
    (movies["revenue"] >= 10000)
][["budget", "revenue"]]
```

```python
# Подсчёт корреляции
corr = data["budget"].corr(data["revenue"])

print(f"Коэффициент корреляции: {corr:.3f}")
```

```python
# График

plt.figure(figsize=(10,6))

plt.xscale("log")
plt.yscale("log")

sns.scatterplot(
    data=data,
    x="budget",
    y="revenue",
    alpha=0.4
)


plt.title("Зависимость кассовых сборов от бюджета фильма")
plt.xlabel("Бюджет")
plt.ylabel("Кассовые сборы")

plt.show()
```

```python
print(data["budget"].describe())
print(data["revenue"].describe())
```

## Итог:
Полученная визуализация показывает положительную зависимость между бюджетом фильма и его кассовыми сборами: фильмы с более высокими бюджетами в среднем собирают большую кассу. При этом наблюдается значительный разброс значений, что свидетельствует о влиянии дополнительных факторов на коммерческий успех фильма.

Коэффициент корреляции составил 0.729, что указывает на сильную положительную связь между бюджетом и кассовыми сборами.

**Вывод**: гипотеза **подтверждается**. Между бюджетом фильма и его кассовыми сборами наблюдается положительная зависимость: увеличение бюджета, как правило, сопровождается ростом кассовых сборов.

## Гипотеза 2: Высокобюджетные фильмы получают более высокие оценки зрителей.

```python
# Подготовка данных
movies["budget"] = pd.to_numeric(movies["budget"], errors="coerce")
movies["vote_average"] = pd.to_numeric(movies["vote_average"], errors="coerce")

data = movies[
    (movies["budget"] > 0) &
    (movies["vote_average"] > 0)
][["budget", "vote_average"]]
```

```python
# Корреляция
corr = data["budget"].corr(data["vote_average"])
print(f"Коэффициент корреляции: {corr:.3f}")
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=data,
    x="budget",
    y="vote_average",
    alpha=0.2
)

plt.xscale("log")

plt.title("Зависимость рейтинга от бюджета фильма")
plt.xlabel("Бюджет")
plt.ylabel("Средняя оценка")

plt.show()
```

Промежуточный итог: из-за большого разброса значений и наличия выбросов значительная часть точек концентрируется в отдельных областях графика, это затрудняет визуальную интерпретацию зависимости между признаками. Для более наглядного анализа применим группировку по диапазонам.

```python
# Создание интервалов
bins = [0, 1e6, 1e7, 5e7, 1e8, float("inf")]

labels = [
    "до 1 млн",
    "1-10 млн",
    "10-50 млн",
    "50-100 млн",
    "свыше 100 млн"
]

data["budget_group"] = pd.cut(
    data["budget"],
    bins=bins,
    labels=labels
)
```

```python
# Группировка
grouped = data.groupby(
    "budget_group"
)["vote_average"].mean()

print(grouped)
```

```python
# Диаграмма
grouped.plot(
    kind="bar",
    figsize=(10,6)
)

plt.title(
    "Средний рейтинг фильмов в зависимости от бюджета"
)

plt.xlabel("Диапазон бюджета")
plt.ylabel("Средний рейтинг")

plt.show()
```

## Итог:

Анализ показал наличие слабой положительной зависимости между бюджетом фильма и его средней оценкой. Несмотря на то, что высокобюджетные фильмы в среднем получают несколько более высокие оценки, различия между группами невелики. Следовательно, бюджет не является определяющим фактором зрительской оценки. **Гипотеза подтверждается лишь частично.**

## Гипотеза 7: Очень популярные фильмы имеют более высокие рейтинги

# Блок 3. Жанры




## Сначала необходимо извлечь жанры.

```python
import ast

def get_genres(text):
    try:
        genres = ast.literal_eval(text)
        return [g["name"] for g in genres]
    except:
        return []

movies["genres_list"] = movies["genres"].apply(
    get_genres
)

genres_df = movies.explode("genres_list")
```

## Гипотеза 8: Жанр Drama получает более высокие оценки, чем средний фильм

```python
# Визуализация
plt.figure(figsize=(12,6))

sns.boxplot(
    data=data.reset_index(drop=True),
    x="genres_list",
    y="vote_average"
)


plt.xticks(rotation=90)

plt.title(
    "Распределение рейтингов по жанрам"
)
plt.xlabel("Жанр")
plt.ylabel("Рейтинг")

plt.show()
```

```python
# Сравнение среднего рейтинга
overall_mean = data["vote_average"].mean()

drama_mean = data[
    data["genres_list"] == "Drama"
]["vote_average"].mean()

print("Средний рейтинг всех фильмов:",
      round(overall_mean, 2))

print("Средний рейтинг Drama:",
      round(drama_mean, 2))
```

## Итог

Boxplot позволяет сравнить распределение рейтингов между жанрами.

Что наблюдаем: медиана и средний рейтинг фильмов жанра Drama обычно находятся выше среднего значения по всей выборке.

Вывод: если средний рейтинг Drama превышает общий средний рейтинг, гипотеза подтверждается.

### Гипотеза 15: Фильмы 90–120 минут получают более высокие оценки, чем очень короткие или длинные


```python
movies['vote_average'] = pd.to_numeric(movies['vote_average'], errors='coerce')
movies['runtime'] = pd.to_numeric(movies['runtime'], errors='coerce')

data = movies[
    (movies['runtime'] >= 40) &
    (movies['runtime'] <= 240) &
    (movies['vote_average'] > 0)
][['runtime', 'vote_average']].copy()
```

```python
bins = [40, 60, 80, 90, 100, 120, 150, 180, 240]
labels = ['40-60', '60-80', '80-90', '90-100', '100-120', '120-150', '150-180', '180-240']

data['runtime_group'] = pd.cut(data['runtime'], bins=bins, labels=labels)
```

```python
grouped = data.groupby('runtime_group', observed=True)['vote_average'].mean()
print(grouped)
```

```python
colors = []
for label in labels:
    if label in ['90-100', '100-120']:
        colors.append('#f4a261')
    else:
        colors.append('#aec6cf')

plt.figure(figsize=(11, 6))
grouped.plot(kind='bar', color=colors)
plt.title('Средний рейтинг фильмов по продолжительности')
plt.xlabel('Продолжительность (мин.)')
plt.ylabel('Средняя оценка')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
target = data[data['runtime_group'].isin(['90-100', '100-120'])]['vote_average'].mean()
others = data[~data['runtime_group'].isin(['90-100', '100-120'])]['vote_average'].mean()

print(f'Средний рейтинг 90-120 мин: {target:.2f}')
print(f'Средний рейтинг остальных: {others:.2f}')
```

## Итог:

Столбчатая диаграмма показывает средний рейтинг по интервалам длительности (оранжевым выделен диапазон 90–120 мин).

Что наблюдаем: фильмы 90–120 мин получают рейтинг ниже, чем короткие (40–60 мин). Длинные фильмы (150–180+ мин) оцениваются ещё выше.

**Вывод:** гипотеза **неверна**. Фильмы 90–120 мин выигрывают у коротких, но проигрывают очень длинным. Зрители склонны высоко оценивать масштабные работы.
