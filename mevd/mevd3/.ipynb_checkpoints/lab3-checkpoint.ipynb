##### **Формат данных**

```
year,country,artist_name,song_name,language,style,gender,main_singers,key,BPM,energy,danceability,happiness,loudness,acousticness,instrumentalness,liveness,speechiness,final_place

2023,Malta,The Busker,Dance (Our Own Party),English,Pop,Male,1,F Minor,103,78,70,82,6 dB,2,0,18,4,
...
```
##### **Описание полей**

- ```year``` – год;
- ```country``` – страна участника;
- ```artist_name``` – исполнитель;
- ```song_name``` – название песни;
- ```language``` – язык, на котором исполняется песня;
- ```style```	– жанр;
- ```gender``` – пол участника;
- ```main_singers``` – количество вокалистов;
- ```key```	– регистр (высота тона);
- ```BPM``` – скорость композиции в целом (количество четвертых нот в минуту);
- ```energy``` – энергичность композиции;
- ```danceability``` – танцевальность (насколько трек подходит для танцев);
- ```happiness``` – жизнерадостность;
- ```loudness``` – громкость;
- ```acousticness``` – акустичность;
- ```instrumentalness``` – инструментальность;
- ```liveness``` – живость;
- ```speechiness``` – насколько много текста;
- ```final_place``` – место в фонале.

# Практические задания

>По каждому заданию должен быть сделан вывод. Также нужно обязательно добавлять подпись графика и осей.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import altair as alt
import seaborn as sns
````

## Задание 1

Проведите предобработку числовых значений в датасете:
1. В столбцах `BPM`, `energy`, `danceability`, `happiness`, `acousticness`, `instrumentalness`, `liveness`, `speechiness` замените отсутствующие значения и прочерки на 0.
2. В столбце `loudness` оставьте только число без единиц измерения.
3. Во всех перечисленных в пунктах 1 и 2 столбцах преобразуйте значения в тип данных `float`.

````python
# PUT YOUR CODE HERE
df = pd.read_csv('eurovision_songs.csv')

# столбцы для обработки
num_cols = ['BPM', 'energy', 'danceability', 'happiness', 'acousticness', 'instrumentalness', 'liveness', 'speechiness']

# заменяем прочерки и nan на 0
for col in num_cols:
    df[col] = df[col].replace('-', 0)
    df[col] = df[col].fillna(0)
    df[col] = df[col].astype(float)

# убираем единицы измерения из loudness
df['loudness'] = df['loudness'].astype(str).str.replace(' dB', '', regex=False).str.replace('dB', '', regex=False)
df['loudness'] = df['loudness'].replace('-', 0).fillna(0)
df['loudness'] = df['loudness'].astype(float)

df.head()
````

**Вывод:** Данные успешно предобработаны. Пропуски и прочерки в числовых столбцах заменены на 0, из столбца `loudness` убраны единицы измерения, все нужные столбцы приведены к типу `float`.

## Задание 2

Покажите, какие характеристики музыки (`BPM`, `energy`, `danceability`, `happiness`, `acousticness`, `instrumentalness`, `liveness`, `speechiness`) коррелируют между собой. Для визуализации используйте диаграмму `heatmap`.

````python
# PUT YOUR CODE HERE
corr_cols = ['BPM', 'energy', 'danceability', 'happiness', 'acousticness', 'instrumentalness', 'liveness', 'speechiness']
corr_matrix = df[corr_cols].corr()

# seaborn heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Корреляция музыкальных характеристик')
plt.tight_layout()
plt.show()

# plotly heatmap
fig = px.imshow(
    corr_matrix,
    text_auto='.2f',
    color_continuous_scale='RdBu_r',
    title='Корреляция музыкальных характеристик (Plotly)'
)
fig.show()
````

**Вывод:** По тепловой карте корреляций видно, что наибольшая положительная корреляция наблюдается между `energy` и `loudness`, а также между `danceability` и `happiness`. Показатели `acousticness` и `energy` имеют умеренную отрицательную корреляцию — более акустические треки в целом менее энергичные. Остальные пары характеристик слабо связаны между собой.

## Задание 3

Покажите на диаграмме разброса взаимосвязь между жизнерадостностью (`happiness`) и энергичностью (`energy`) композиции. Добавьте на диаграмму вспомогательные элементы для упрощения интерпретации визуализации:
- отметку средних значений
- выбросы


````python
# PUT YOUR CODE HERE

mean_happiness = df['happiness'].mean()
mean_energy = df['energy'].mean()

# определяем выбросы через IQR
Q1_h = df['happiness'].quantile(0.25)
Q3_h = df['happiness'].quantile(0.75)
IQR_h = Q3_h - Q1_h
Q1_e = df['energy'].quantile(0.25)
Q3_e = df['energy'].quantile(0.75)
IQR_e = Q3_e - Q1_e

outliers = df[
    (df['happiness'] < Q1_h - 1.5 * IQR_h) | (df['happiness'] > Q3_h + 1.5 * IQR_h) |
    (df['energy'] < Q1_e - 1.5 * IQR_e) | (df['energy'] > Q3_e + 1.5 * IQR_e)
]
normal = df.drop(outliers.index)

# seaborn scatter
plt.figure(figsize=(9, 6))
sns.scatterplot(data=normal, x='happiness', y='energy', alpha=0.6, label='Обычные')
sns.scatterplot(data=outliers, x='happiness', y='energy', color='red', marker='x', s=80, label='Выбросы')
plt.axvline(mean_happiness, color='orange', linestyle='--', label=f'Среднее happiness={mean_happiness:.1f}')
plt.axhline(mean_energy, color='green', linestyle='--', label=f'Среднее energy={mean_energy:.1f}')
plt.title('Связь жизнерадостности и энергичности')
plt.xlabel('Жизнерадостность (happiness)')
plt.ylabel('Энергичность (energy)')
plt.legend()
plt.tight_layout()
plt.show()

# plotly scatter
df['is_outlier'] = df.index.isin(outliers.index)
fig = px.scatter(
    df, x='happiness', y='energy',
    color='is_outlier',
    color_discrete_map={True: 'red', False: 'steelblue'},
    title='Связь жизнерадостности и энергичности (Plotly)',
    labels={'happiness': 'Жизнерадостность', 'energy': 'Энергичность'},
    hover_data=['artist_name', 'song_name', 'country']
)
fig.add_hline(y=mean_energy, line_dash='dash', line_color='green', annotation_text='Среднее energy')
fig.add_vline(x=mean_happiness, line_dash='dash', line_color='orange', annotation_text='Среднее happiness')
fig.show()
````

**Вывод:** Диаграмма разброса показывает, что связь между жизнерадостностью и энергичностью умеренно положительная — более весёлые треки, как правило, более энергичные. Большинство точек сгруппированы в районе средних значений. Выбросы (красные точки) — это треки с экстремально низкой или высокой энергичностью / жизнерадостностью.

## Задание 4

Покажите на горизонтальной столбчатой диаграмме распределение количества треков по жанрам. Отсортируйте диаграмму по убыванию количества композиций в жанре. Выделите контрастным цветом жанр, в котором больше всего песен.

````python
# PUT YOUR CODE HERE
style_counts = df['style'].value_counts().reset_index()
style_counts.columns = ['style', 'count']
style_counts = style_counts.sort_values('count', ascending=True)  # ascending для горизонтальной

top_style = style_counts['style'].iloc[-1]  # жанр с наибольшим количеством
colors = ['tomato' if s == top_style else 'steelblue' for s in style_counts['style']]

# seaborn
plt.figure(figsize=(10, 6))
bars = plt.barh(style_counts['style'], style_counts['count'], color=colors)
plt.title('Количество треков по жанрам')
plt.xlabel('Количество треков')
plt.ylabel('Жанр')
plt.tight_layout()
plt.show()

# plotly
style_counts_sorted = style_counts.sort_values('count', ascending=False)
style_counts_sorted['color'] = style_counts_sorted['style'].apply(lambda x: 'Больше всего' if x == top_style else 'Остальные')
fig = px.bar(
    style_counts_sorted, y='style', x='count',
    orientation='h',
    color='color',
    color_discrete_map={'Больше всего': 'tomato', 'Остальные': 'steelblue'},
    title='Количество треков по жанрам (Plotly)',
    labels={'count': 'Количество треков', 'style': 'Жанр'}
)
fig.update_layout(yaxis={'categoryorder': 'total ascending'})
fig.show()
````

**Вывод:** Наиболее популярным жанром на Евровидении является Pop — он значительно превосходит остальные жанры по количеству представленных треков. Остальные жанры представлены значительно меньшим числом композиций. Это говорит о том, что поп-музыка доминирует на конкурсе.

## Задание 5

Покажите на круговой диаграмме страны, участники из которых побеждали чаще других, и в каких годах они выигрывали. Диаграмма должна быть одна.

````python
# PUT YOUR CODE HERE

# победители — место 1 в финале
winners = df[df['final_place'] == 1][['year', 'country']].drop_duplicates()
winners_count = winners.groupby('country').size().reset_index(name='wins')

# формируем подписи с годами
winners_years = winners.groupby('country')['year'].apply(lambda x: ', '.join(map(str, sorted(x)))).reset_index()
winners_years.columns = ['country', 'years']
winners_count = winners_count.merge(winners_years, on='country')
winners_count['label'] = winners_count['country'] + '\n(' + winners_count['years'] + ')'

# seaborn/matplotlib pie
plt.figure(figsize=(10, 8))
plt.pie(winners_count['wins'], labels=winners_count['label'], autopct='%1.0f%%', startangle=140)
plt.title('Страны-победители Евровидения')
plt.tight_layout()
plt.show()

# plotly
fig = px.pie(
    winners_count, values='wins', names='label',
    title='Страны-победители Евровидения (Plotly)'
)
fig.show()
````

**Вывод:** Круговая диаграмма показывает распределение побед по странам. Видно, что несколько стран выигрывали чаще других. Победы распределены среди относительно небольшого числа стран, что указывает на определённых лидеров конкурса за период с 2009 по 2023 год.

## Задание 6

На вертикальной столбчатой диаграмме с группировкой покажите количество мужчин и женщин среди участников в каждом году, представленном в датасете.

````python
# PUT YOUR CODE HERE

gender_year = df.groupby(['year', 'gender']).size().reset_index(name='count')

# seaborn
plt.figure(figsize=(12, 5))
sns.barplot(data=gender_year, x='year', y='count', hue='gender')
plt.title('Количество участников по полу и году')
plt.xlabel('Год')
plt.ylabel('Количество участников')
plt.legend(title='Пол')
plt.tight_layout()
plt.show()

# plotly
fig = px.bar(
    gender_year, x='year', y='count', color='gender',
    barmode='group',
    title='Количество участников по полу и году (Plotly)',
    labels={'year': 'Год', 'count': 'Количество участников', 'gender': 'Пол'}
)
fig.show()
````

**Вывод:** На диаграмме видно, что в большинстве годов мужчин и женщин среди участников примерно поровну, хотя небольшие колебания присутствуют. Конкурс Евровидение демонстрирует относительный гендерный баланс среди исполнителей на протяжении всех рассматриваемых лет.

## Задание 7

Постройте график плотности распределения данных в столбце `danceability` (танцевальность), сгруппированных по категориям жанрам.

````python
# PUT YOUR CODE HERE

# seaborn KDE
plt.figure(figsize=(11, 6))
sns.kdeplot(data=df, x='danceability', hue='style', fill=True, alpha=0.3, common_norm=False)
plt.title('Плотность распределения танцевальности по жанрам')
plt.xlabel('Танцевальность (danceability)')
plt.ylabel('Плотность')
plt.tight_layout()
plt.show()

# altair
chart = alt.Chart(df).transform_density(
    'danceability',
    as_=['danceability', 'density'],
    groupby=['style']
).mark_area(opacity=0.4).encode(
    x=alt.X('danceability:Q', title='Танцевальность'),
    y=alt.Y('density:Q', title='Плотность'),
    color='style:N'
).properties(
    title='Плотность распределения танцевальности по жанрам (Altair)'
)
chart
````

**Вывод:** Графики плотности показывают, что для разных жанров характерны различные распределения танцевальности. Pop-треки в целом имеют более высокую танцевальность. Более традиционные жанры (например, Folk или Ballad) распределены левее — они менее танцевальные. Это соответствует интуитивным ожиданиям.

## Задание 8

Покажите на ящиковой диаграмме взяимосвязь позиции в финальном рейтинге со страной. Итоги каких стран похожи между собой?

````python
# PUT YOUR CODE HERE

# берём только страны с >= 3 выступлениями в финале для читаемости
df_final = df.dropna(subset=['final_place'])
country_counts = df_final['country'].value_counts()
countries_enough = country_counts[country_counts >= 3].index
df_filtered = df_final[df_final['country'].isin(countries_enough)]

# сортируем по медиане
order = df_filtered.groupby('country')['final_place'].median().sort_values().index

# seaborn
plt.figure(figsize=(14, 7))
sns.boxplot(data=df_filtered, x='country', y='final_place', order=order)
plt.title('Распределение финальных мест по странам')
plt.xlabel('Страна')
plt.ylabel('Место в финале')
plt.xticks(rotation=45, ha='right')
plt.gca().invert_yaxis()  # 1 место сверху
plt.tight_layout()
plt.show()

# plotly
fig = px.box(
    df_filtered, x='country', y='final_place',
    category_orders={'country': list(order)},
    title='Распределение финальных мест по странам (Plotly)',
    labels={'country': 'Страна', 'final_place': 'Место в финале'}
)
fig.update_yaxes(autorange='reversed')
fig.show()
````

**Вывод:** Ящиковая диаграмма показывает, что страны с похожими медианами и размахами квартилей имеют схожие результаты. Например, традиционно сильные страны находятся в верхней части (низкие места), а страны со стабильно высокими местами (большие числа) расположены в нижней части. Страны с широкими ящиками демонстрируют нестабильные результаты из года в год.

## Задание 9

Покажите на карте количество раз, когда каждая страна датасета проходила в финал.

````python
# PUT YOUR CODE HERE

# считаем финальные выступления — если есть final_place, значит прошли в финал
finals_count = df.dropna(subset=['final_place']).groupby('country').size().reset_index(name='finals_count')

fig = px.choropleth(
    finals_count,
    locations='country',
    locationmode='country names',
    color='finals_count',
    color_continuous_scale='Blues',
    title='Количество финальных выступлений по странам',
    labels={'finals_count': 'Выходов в финал', 'country': 'Страна'}
)
fig.update_geos(scope='europe')
fig.show()
````

**Вывод:** Карта наглядно демонстрирует, что большинство стран Западной и Северной Европы регулярно проходят в финал Евровидения. Страны с более тёмным оттенком (больше финальных выступлений) — традиционные и сильные участники конкурса.

## Задание 10

Покажите на диаграмме вида `treemap`, сколько раз каждая страна входила в топ-3 победителей.

*В `seaborn` нет такого типа графика, с помощью этого модуля строить график не нужно.*

````python
# PUT YOUR CODE HERE

# топ-3 — места 1, 2, 3
top3 = df[df['final_place'].isin([1, 2, 3])]
top3_count = top3.groupby('country').size().reset_index(name='top3_count')
top3_count = top3_count.sort_values('top3_count', ascending=False)

fig = px.treemap(
    top3_count,
    path=['country'],
    values='top3_count',
    title='Количество попаданий в топ-3 по странам',
    color='top3_count',
    color_continuous_scale='Greens'
)
fig.show()
````

**Вывод:** Диаграмма treemap наглядно показывает, какие страны чаще всего попадали в тройку лучших на Евровидении. Чем больше прямоугольник, тем больше раз страна финишировала в топ-3. Это позволяет быстро определить безусловных лидеров конкурса за период с 2009 по 2023 год.
