# Данные

По [ссылке](https://drive.google.com/drive/folders/1e0E4vQzaE9LBZgfzz0AzsyAXgtH0eMaP?usp=share_link) расположена таблица с информацией о песнях, исполненных на Евровидении с 2009 по 2023 год. Все задания лабораторной работы необходимо выполнять по этим данным. Данные в полном объеме можно найти [по ссылке](https://www.kaggle.com/datasets/diamondsnake/eurovision-song-contest-data).

##### **Формат данных**

```
year,country,artist_name,song_name,language,style,gender,main_singers,key,BPM,energy,danceability,happiness,loudness,acousticness,instrumentalness,liveness,speechiness,final_place
2023,Norway,Alessandra,Queen of Kings,English,Pop,Female,1,E Minor,110,36,64,23,10 dB,58,0,10,3,5
2023,Malta,The Busker,Dance (Our Own Party),English,Pop,Male,1,F Minor,103,78,70,82,6 dB,2,0,18,4,
...
```
##### **Описание полей**

- `year` – год;
- `country` – страна участника;
- `artist_name` – исполнитель;
- `song_name` – название песни;
- `language` – язык, на котором исполняется песня;
- `style` – жанр;
- `gender` – пол участника;
- `main_singers` – количество вокалистов;
- `key` – регистр (высота тона);
- `BPM` – скорость композиции в целом (количество четвертых нот в минуту);
- `energy` – энергичность композиции;
- `danceability` – танцевальность (насколько трек подходит для танцев);
- `happiness` – жизнерадостность;
- `loudness` – громкость;
- `acousticness` – акустичность;
- `instrumentalness` – инструментальность;
- `liveness` – живость;
- `speechiness` – насколько много текста;
- `final_place` – место в финале.


# Практические задания

> По каждому заданию должен быть сделан вывод. Также нужно обязательно добавлять подпись графика и осей.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import altair as alt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
````

## Загрузка данных

````python
df = pd.read_csv('song_data.csv')
df.head()
````

## Задание 1

Проведите предобработку числовых значений в датасете:
1. В столбцах `BPM`, `energy`, `danceability`, `happiness`, `acousticness`, `instrumentalness`, `liveness`, `speechiness` замените отсутствующие значения и прочерки на 0.
2. В столбце `loudness` оставьте только число без единиц измерения.
3. Во всех перечисленных в пунктах 1 и 2 столбцах преобразуйте значения в тип данных `float`.


````python
# столбцы которые надо почистить
num_cols = ['BPM', 'energy', 'danceability', 'happiness',
            'acousticness', 'instrumentalness', 'liveness', 'speechiness']

# заменяем прочерки и пропуски на 0
for col in num_cols:
    df[col] = df[col].replace('-', 0)
    df[col] = df[col].fillna(0)
    df[col] = df[col].astype(float)

# убираем ' dB' из loudness
df['loudness'] = df['loudness'].str.replace(' dB', '', regex=False)
df['loudness'] = df['loudness'].fillna(0).astype(float)

df[num_cols + ['loudness']].dtypes
````

**Вывод:** Предобработка выполнена успешно. Все нужные столбцы теперь имеют тип `float`, пропуски и прочерки заменены на 0, единицы измерения убраны из `loudness`.

## Задание 2

Покажите, какие характеристики музыки (`BPM`, `energy`, `danceability`, `happiness`, `acousticness`, `instrumentalness`, `liveness`, `speechiness`) коррелируют между собой. Для визуализации используйте диаграмму `heatmap`.


````python
corr_cols = ['BPM', 'energy', 'danceability', 'happiness',
             'acousticness', 'instrumentalness', 'liveness', 'speechiness']

corr_matrix = df[corr_cols].corr()

# seaborn heatmap
plt.figure(figsize=(9, 7))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Корреляция между музыкальными характеристиками')
plt.tight_layout()
plt.show()
````

````python
# plotly heatmap
fig = px.imshow(
    corr_matrix,
    text_auto='.2f',
    color_continuous_scale='RdBu_r',
    title='Корреляция между музыкальными характеристиками (Plotly)'
)
fig.show()
````

**Вывод:** Наибольшая положительная корреляция наблюдается между `energy` и `loudness`, а также `energy` и `happiness`. Характеристики `acousticness` и `energy` имеют отрицательную корреляцию, что логично: акустические треки обычно менее энергичны.

## Задание 3

Покажите на диаграмме разброса взаимосвязь между жизнерадостностью (`happiness`) и энергичностью (`energy`) композиции. Добавьте на диаграмму вспомогательные элементы для упрощения интерпретации визуализации:
- отметку средних значений
- выбросы


````python
mean_h = df['happiness'].mean()
mean_e = df['energy'].mean()

# определяем выбросы через IQR
Q1_h = df['happiness'].quantile(0.25)
Q3_h = df['happiness'].quantile(0.75)
IQR_h = Q3_h - Q1_h

Q1_e = df['energy'].quantile(0.25)
Q3_e = df['energy'].quantile(0.75)
IQR_e = Q3_e - Q1_e

is_outlier = (
    (df['happiness'] < Q1_h - 1.5 * IQR_h) | (df['happiness'] > Q3_h + 1.5 * IQR_h) |
    (df['energy'] < Q1_e - 1.5 * IQR_e) | (df['energy'] > Q3_e + 1.5 * IQR_e)
)

df_normal = df[~is_outlier]
df_outliers = df[is_outlier]

# seaborn
plt.figure(figsize=(9, 6))
sns.scatterplot(data=df_normal, x='happiness', y='energy', alpha=0.6, label='Обычные точки')
sns.scatterplot(data=df_outliers, x='happiness', y='energy',
                color='red', marker='X', s=80, label='Выбросы')
plt.axvline(mean_h, color='blue', linestyle='--', alpha=0.7, label=f'Среднее happiness={mean_h:.1f}')
plt.axhline(mean_e, color='orange', linestyle='--', alpha=0.7, label=f'Среднее energy={mean_e:.1f}')
plt.title('Взаимосвязь жизнерадостности и энергичности')
plt.xlabel('Жизнерадостность (happiness)')
plt.ylabel('Энергичность (energy)')
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# plotly
df_plot = df.copy()
df_plot['outlier'] = is_outlier.map({True: 'Выброс', False: 'Обычная точка'})

fig = px.scatter(
    df_plot, x='happiness', y='energy',
    color='outlier',
    color_discrete_map={'Выброс': 'red', 'Обычная точка': '#636EFA'},
    title='Взаимосвязь жизнерадостности и энергичности (Plotly)',
    labels={'happiness': 'Жизнерадостность', 'energy': 'Энергичность'},
    hover_data=['artist_name', 'song_name']
)
fig.add_hline(y=mean_e, line_dash='dash', line_color='orange',
              annotation_text=f'Среднее energy={mean_e:.1f}')
fig.add_vline(x=mean_h, line_dash='dash', line_color='blue',
              annotation_text=f'Среднее happiness={mean_h:.1f}')
fig.show()
````

**Вывод:** Прослеживается слабая положительная взаимосвязь — более энергичные треки чаще оказываются и более жизнерадостными. Выбросы в основном сосредоточены в зоне очень низкой жизнерадостности при высокой энергии (тяжёлые жанры).

## Задание 4

Покажите на горизонтальной столбчатой диаграмме распределение количества треков по жанрам. Отсортируйте диаграмму по убыванию количества композиций в жанре. Выделите контрастным цветом жанр, в котором больше всего песен.


````python
style_counts = df['style'].value_counts().reset_index()
style_counts.columns = ['style', 'count']
style_counts = style_counts.sort_values('count', ascending=True)  # ascending для горизонтального

# цвета: самый популярный — выделен
max_style = style_counts['style'].iloc[-1]
colors = ['#ff6b6b' if s == max_style else '#74b9ff' for s in style_counts['style']]

# seaborn
plt.figure(figsize=(9, 5))
bars = plt.barh(style_counts['style'], style_counts['count'], color=colors)
plt.xlabel('Количество треков')
plt.ylabel('Жанр')
plt.title('Количество треков по жанрам')
for bar, val in zip(bars, style_counts['count']):
    plt.text(bar.get_width() + 2, bar.get_y() + bar.get_height()/2,
             str(val), va='center')
plt.tight_layout()
plt.show()
````

````python
# plotly
style_counts_sorted = style_counts.sort_values('count', ascending=False)
style_counts_sorted['color'] = style_counts_sorted['style'].apply(
    lambda x: 'Наибольшее количество' if x == max_style else 'Остальные жанры'
)

fig = px.bar(
    style_counts_sorted, y='style', x='count',
    orientation='h',
    color='color',
    color_discrete_map={'Наибольшее количество': '#ff6b6b', 'Остальные жанры': '#74b9ff'},
    title='Количество треков по жанрам (Plotly)',
    labels={'count': 'Количество треков', 'style': 'Жанр'}
)
fig.update_layout(yaxis={'categoryorder': 'total ascending'})
fig.show()
````

**Вывод:** Жанр **Pop** является самым популярным на Евровидении — более половины всех участников исполняют поп-музыку. Реже всего встречается жанр Opera.

## Задание 5

Покажите на круговой диаграмме страны, участники из которых побеждали чаще других, и в каких годах они выигрывали. Диаграмма должна быть одна.


````python
winners = df[df['final_place'] == 1.0][['year', 'country']].copy()
winners['label'] = winners['country'] + ' (' + winners['year'].astype(str) + ')'

# seaborn/matplotlib — круговая
fig, ax = plt.subplots(figsize=(10, 8))
ax.pie(
    [1] * len(winners),
    labels=winners['label'],
    autopct=lambda p: '',
    startangle=140,
    pctdistance=0.85
)
ax.set_title('Страны-победители Евровидения и годы побед')
plt.tight_layout()
plt.show()
````

````python
# plotly — каждый победитель как отдельный сектор
fig = px.pie(
    winners,
    names='label',
    title='Страны-победители Евровидения (2009–2023)',
    hole=0.3
)
fig.update_traces(textposition='inside', textinfo='label')
fig.show()
````

**Вывод:** Швеция побеждала чаще всех — 3 раза за рассматриваемый период (2012, 2015, 2023). Украина также побеждала дважды (2016, 2022). Остальные страны-победители выигрывали по одному разу.

## Задание 6

На вертикальной столбчатой диаграмме с группировкой покажите количество мужчин и женщин среди участников в каждом году, представленном в датасете.


````python
# фильтруем только Male и Female, Mix убираем
gender_year = df[df['gender'].isin(['Male', 'Female'])].groupby(['year', 'gender']).size().reset_index(name='count')

# seaborn
plt.figure(figsize=(12, 5))
sns.barplot(data=gender_year, x='year', y='count', hue='gender',
            palette={'Female': '#ff6b9d', 'Male': '#4facfe'})
plt.title('Количество участников по полу в каждом году')
plt.xlabel('Год')
plt.ylabel('Количество участников')
plt.legend(title='Пол')
plt.tight_layout()
plt.show()
````

````python
# plotly
fig = px.bar(
    gender_year, x='year', y='count', color='gender',
    barmode='group',
    color_discrete_map={'Female': '#ff6b9d', 'Male': '#4facfe'},
    title='Количество участников по полу в каждом году (Plotly)',
    labels={'count': 'Количество участников', 'year': 'Год', 'gender': 'Пол'}
)
fig.show()
````

**Вывод:** В большинстве лет количество женщин-участниц немного превышает количество мужчин. Заметных резких изменений в гендерном распределении участников по годам не наблюдается.

## Задание 7

Постройте график плотности распределения данных в столбце `danceability` (танцевальность), сгруппированных по категориям жанрам.


````python
# seaborn kdeplot
plt.figure(figsize=(10, 6))
for style in df['style'].unique():
    subset = df[df['style'] == style]['danceability']
    sns.kdeplot(subset, label=style, fill=False)
plt.title('Плотность распределения танцевальности по жанрам')
plt.xlabel('Danceability')
plt.ylabel('Плотность')
plt.legend(title='Жанр')
plt.tight_layout()
plt.show()
````

````python
# altair
chart = alt.Chart(df).transform_density(
    'danceability',
    as_=['danceability', 'density'],
    groupby=['style']
).mark_line().encode(
    x=alt.X('danceability:Q', title='Танцевальность'),
    y=alt.Y('density:Q', title='Плотность'),
    color=alt.Color('style:N', title='Жанр')
).properties(
    title='Плотность распределения танцевальности по жанрам (Altair)',
    width=600,
    height=350
)
chart
````

**Вывод:** Жанр Dance ожидаемо имеет наивысшую медианную танцевальность. Ballad и Opera сосредоточены в нижней части шкалы танцевальности. Pop занимает широкий диапазон значений.

## Задание 8

Покажите на ящиковой диаграмме взаимосвязь позиции в финальном рейтинге со страной. Итоги каких стран похожи между собой?


````python
# берём страны с достаточным количеством финальных выступлений
country_counts = df.dropna(subset=['final_place'])['country'].value_counts()
top_countries = country_counts[country_counts >= 7].index.tolist()

df_top = df[df['country'].isin(top_countries)].dropna(subset=['final_place'])

# сортируем по медиане
order = df_top.groupby('country')['final_place'].median().sort_values().index

plt.figure(figsize=(14, 7))
sns.boxplot(data=df_top, x='country', y='final_place', order=order, palette='Set3')
plt.title('Распределение финальных мест по странам')
plt.xlabel('Страна')
plt.ylabel('Место в финале')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

````python
# plotly
fig = px.box(
    df_top, x='country', y='final_place',
    category_orders={'country': list(order)},
    title='Распределение финальных мест по странам (Plotly)',
    labels={'final_place': 'Место в финале', 'country': 'Страна'}
)
fig.update_layout(xaxis_tickangle=-45)
fig.show()
````

**Вывод:** Швеция и Украина стабильно показывают высокие результаты — их ящики смещены вниз (ближе к 1-му месту). Страны вроде Испании и Великобритании имеют широкий разброс результатов. Похожие результаты у Норвегии и Дании — обе стабильно держатся в середине таблицы.

## Задание 9

Покажите на карте количество раз, когда каждая страна датасета проходила в финал.


````python
# страна попала в финал, если final_place не NaN
finals_count = df.dropna(subset=['final_place']).groupby('country').size().reset_index(name='finals')

# seaborn не поддерживает карты, используем plotly choropleth
# через matplotlib + geopandas сложно, поэтому для seaborn покажем столбчатую как альтернативу
finals_sorted = finals_count.sort_values('finals', ascending=False)

plt.figure(figsize=(14, 6))
sns.barplot(data=finals_sorted, x='country', y='finals', palette='Blues_d')
plt.title('Количество попаданий в финал по странам')
plt.xlabel('Страна')
plt.ylabel('Количество финалов')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
````

````python
# plotly choropleth карта
fig = px.choropleth(
    finals_count,
    locations='country',
    locationmode='country names',
    color='finals',
    color_continuous_scale='Blues',
    scope='europe',
    title='Количество попаданий в финал Евровидения по странам',
    labels={'finals': 'Кол-во финалов'}
)
fig.show()
````

**Вывод:** Постоянные финалисты — Испания, Германия, Франция и Великобритания (Big Five, гарантированный выход в финал). Среди остальных стран по числу финальных выходов лидируют Швеция, Украина и Норвегия.

## Задание 10

Покажите на диаграмме вида `treemap`, сколько раз каждая страна входила в топ-3 победителей.

*В `seaborn` нет такого типа графика, с помощью этого модуля строить график не нужно.*


````python
top3 = df[df['final_place'] <= 3].groupby('country').size().reset_index(name='top3_count')
top3 = top3.sort_values('top3_count', ascending=False)

fig = px.treemap(
    top3,
    path=['country'],
    values='top3_count',
    color='top3_count',
    color_continuous_scale='Greens',
    title='Количество попаданий в топ-3 по странам'
)
fig.update_traces(textinfo='label+value')
fig.show()
````

**Вывод:** Швеция чаще всего попадала в топ-3 победителей Евровидения за рассматриваемый период. Также высокие позиции у Италии, России (до 2022 года) и Украины. Это подтверждает, что именно эти страны стабильно показывают высокие результаты на конкурсе.
