# РГР №3 -- Регрессионные модели и их интерпретация

**Вариант:** A-7  
**Предметная область:** Теплопакет (TDP, Вт) процессоров AMD в зависимости от числа ядер  
**Источник данных:** `openintro::cpu` -- набор данных о процессорах, выпущенных с 2010 по 2020 год  
**Ссылка:** https://vincentarelbundock.github.io/Rdatasets/doc/openintro/cpu.html  
**Срез:** фильтр `company == "AMD"`, итого 60 наблюдений  

**Переменные:**
- `x` -- число ядер (`cores`)
- `y` -- TDP, Вт (`tdp`)

**Дополнительная модель по варианту:** степенная -- `y = a * x^b`  
**Прогнозное значение:** `x* = 9.3280` (106% от среднего числа ядер)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# грузим данные
df = pd.read_csv('RGR3_A_7.csv')
x = df['x'].values
y = df['y'].values

print(f'Загружено {len(df)} наблюдений')
print(df.describe())
```

## 1. Первичный анализ данных

Посмотрим на данные визуально и сформулируем гипотезу о характере связи между числом ядер и TDP.

```python
# диаграмма рассеяния
plt.figure(figsize=(8, 5))
plt.scatter(x, y, alpha=0.6, color='steelblue', edgecolors='k', linewidths=0.4)
plt.xlabel('Число ядер (x)')
plt.ylabel('TDP, Вт (y)')
plt.title('Диаграмма рассеяния: TDP от числа ядер (AMD)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

print(f'Среднее x: {np.mean(x):.4f}')
print(f'Среднее y: {np.mean(y):.4f}')
print(f'x*: 9.3280')
```

**Гипотеза о характере связи:**

На диаграмме рассеяния видно, что с ростом числа ядер TDP в целом увеличивается -- связь положительная. При малом числе ядер (2-4) разброс значений TDP очень большой (от ~9 до ~125 Вт), что объясняется широкой линейкой процессоров с разными классами энергопотребления (мобильные, настольные, серверные). При большом числе ядер (16+) значения TDP концентрируются в диапазоне 115-280 Вт.

Зависимость выглядит скорее нелинейной -- похоже на степенной или логарифмический рост. Попробуем линейную, квадратичную и степенную модели.

## 2. Построение трёх моделей

### 2.1 Линейная модель: y = a + b*x

```python
n = len(x)
x_mean = np.mean(x)
y_mean = np.mean(y)

# оценки МНК вручную (как в методичке)
Sxx = np.sum((x - x_mean)**2)
Sxy = np.sum((x - x_mean) * (y - y_mean))

b_lin = Sxy / Sxx
a_lin = y_mean - b_lin * x_mean

y_pred_lin = a_lin + b_lin * x

print(f'Линейная модель: y = {a_lin:.4f} + {b_lin:.4f} * x')
```

### 2.2 Квадратичная модель: y = a + b*x + c*x^2

```python
# используем numpy polyfit -- удобно и быстро
coeffs_quad = np.polyfit(x, y, 2)
c_q, b_q, a_q = coeffs_quad

y_pred_quad = np.polyval(coeffs_quad, x)

print(f'Квадратичная модель: y = {a_q:.4f} + {b_q:.4f}*x + {c_q:.4f}*x^2')
```

### 2.3 Степенная модель: y = a * x^b

Линеаризуем: ln(y) = ln(a) + b*ln(x), то есть делаем замену Y = ln(y), X = ln(x) и строим линейную регрессию Y на X.

```python
# логарифмируем -- все значения x и y положительные, проблем нет
ln_x = np.log(x)
ln_y = np.log(y)

ln_x_mean = np.mean(ln_x)
ln_y_mean = np.mean(ln_y)

Sxx_pow = np.sum((ln_x - ln_x_mean)**2)
Sxy_pow = np.sum((ln_x - ln_x_mean) * (ln_y - ln_y_mean))

b_pow = Sxy_pow / Sxx_pow
ln_a_pow = ln_y_mean - b_pow * ln_x_mean
a_pow = np.exp(ln_a_pow)

y_pred_pow = a_pow * x**b_pow

print(f'Степенная модель: y = {a_pow:.4f} * x^{b_pow:.4f}')
```

### Графики всех трёх моделей

```python
x_plot = np.linspace(min(x), max(x), 300)
y_lin_plot = a_lin + b_lin * x_plot
y_quad_plot = np.polyval(coeffs_quad, x_plot)
y_pow_plot = a_pow * x_plot**b_pow

plt.figure(figsize=(9, 6))
plt.scatter(x, y, alpha=0.6, color='steelblue', edgecolors='k', linewidths=0.4, label='Данные', zorder=5)
plt.plot(x_plot, y_lin_plot, 'r-', linewidth=2, label=f'Линейная: y={a_lin:.2f}+{b_lin:.2f}x')
plt.plot(x_plot, y_quad_plot, 'g--', linewidth=2, label=f'Квадратичная')
plt.plot(x_plot, y_pow_plot, 'm-.', linewidth=2, label=f'Степенная: y={a_pow:.2f}*x^{b_pow:.2f}')
plt.xlabel('Число ядер (x)')
plt.ylabel('TDP, Вт (y)')
plt.title('Сравнение регрессионных моделей')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## 3. Сравнение моделей

```python
def r2_score(y_true, y_pred):
    TSS = np.sum((y_true - np.mean(y_true))**2)
    RSS = np.sum((y_true - y_pred)**2)
    return 1 - RSS / TSS

def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred)**2))

def mape(y_true, y_pred):
    # средняя относительная ошибка аппроксимации (в %)
    return 100 / len(y_true) * np.sum(np.abs((y_true - y_pred) / y_true))

models = {
    'Линейная':      y_pred_lin,
    'Квадратичная':  y_pred_quad,
    'Степенная':     y_pred_pow
}

print(f'{"Модель":<16} {"R^2":>8} {"RMSE":>8} {"A, %":>8}')
print('-' * 44)
for name, y_pred in models.items():
    r2  = r2_score(y, y_pred)
    rm  = rmse(y, y_pred)
    a   = mape(y, y_pred)
    print(f'{name:<16} {r2:>8.4f} {rm:>8.2f} {a:>8.2f}')
```

**Вывод по сравнению моделей:**

Все три модели дают схожий уровень R^2 (около 0.55-0.65). Высокий разброс TDP при фиксированном числе ядер объясняется тем, что у AMD с одинаковым числом ядер выпускаются процессоры разных классов: мобильные (низкий TDP), настольные и серверные (высокий TDP). Степенная и квадратичная модели чуть лучше захватывают нелинейный рост TDP при переходе к многоядерным процессорам. Степенная модель содержательно оправдана: энергопотребление растёт как степенная функция от числа ядер.

Для подробного статистического анализа по условию задания используем линейную модель.

### Графики остатков

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
titles = ['Линейная', 'Квадратичная', 'Степенная']
preds  = [y_pred_lin, y_pred_quad, y_pred_pow]

for ax, title, yp in zip(axes, titles, preds):
    residuals = y - yp
    ax.scatter(yp, residuals, alpha=0.6, color='steelblue', edgecolors='k', linewidths=0.3)
    ax.axhline(0, color='red', linestyle='--')
    ax.set_xlabel('y_hat')
    ax.set_ylabel('Остаток')
    ax.set_title(title)
    ax.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

## 4. Подробный статистический анализ линейной модели

### 4.1 Коэффициенты, остатки, RSS, дисперсия ошибки

```python
e = y - y_pred_lin  # остатки
RSS = np.sum(e**2)
s2  = RSS / (n - 2)
s   = np.sqrt(s2)

print(f'theta_0 (a) = {a_lin:.4f}')
print(f'theta_1 (b) = {b_lin:.4f}')
print(f'RSS         = {RSS:.4f}')
print(f's^2         = {s2:.4f}')
print(f's           = {s:.4f}')
```

### 4.2 Стандартные ошибки коэффициентов и 95% доверительные интервалы

```python
SE_b = s / np.sqrt(Sxx)
SE_a = s * np.sqrt(1/n + x_mean**2 / Sxx)

# квантиль t_{0.975; n-2}
t_crit = stats.t.ppf(0.975, df=n-2)

CI_b = (b_lin - t_crit * SE_b, b_lin + t_crit * SE_b)
CI_a = (a_lin - t_crit * SE_a, a_lin + t_crit * SE_a)

print(f'SE(theta_1) = {SE_b:.4f}')
print(f'SE(theta_0) = {SE_a:.4f}')
print(f't_крит      = {t_crit:.4f}')
print()
print(f'95% ДИ для theta_1 (b): [{CI_b[0]:.4f}, {CI_b[1]:.4f}]')
print(f'95% ДИ для theta_0 (a): [{CI_a[0]:.4f}, {CI_a[1]:.4f}]')
```

### 4.3 Проверка гипотезы H0: theta_1 = 0 (значимость наклона)

```python
t_obs = b_lin / SE_b
p_value = 2 * (1 - stats.t.cdf(abs(t_obs), df=n-2))

print(f't_набл  = {t_obs:.4f}')
print(f't_крит  = {t_crit:.4f}')
print(f'p-value = {p_value:.6f}')
print()
if abs(t_obs) > t_crit:
    print('|t_набл| > t_крит => H0 отвергается, коэффициент при x значим на уровне 0.05')
else:
    print('|t_набл| <= t_крит => H0 не отвергается')
```

**Вывод:** Нулевая гипотеза о незначимости коэффициента наклона отвергается -- линейная зависимость между числом ядер и TDP статистически значима. Доверительный интервал для theta_1 не содержит ноль, что подтверждает вывод. p-value крайне мало.

При этом R^2 не очень высокий (~0.57), что говорит о том, что число ядер объясняет меньше половины дисперсии TDP -- оставшийся разброс обусловлен классом процессора (мобильный vs настольный vs серверный), частотой и другими характеристиками.

### 4.4 Таблица остатков (первые 10 наблюдений)

```python
res_df = pd.DataFrame({
    'x':     x,
    'y':     y,
    'y_hat': np.round(y_pred_lin, 2),
    'e':     np.round(e, 2)
})
print(res_df.head(10).to_string(index=False))
```

## 5. Прогноз при x* = 9.3280

```python
x_star = 9.3280

y_star_lin  = a_lin + b_lin * x_star
y_star_quad = np.polyval(coeffs_quad, x_star)
y_star_pow  = a_pow * x_star**b_pow

print(f'Прогноз при x* = {x_star}:')
print(f'  Линейная:     y_hat = {y_star_lin:.2f} Вт')
print(f'  Квадратичная: y_hat = {y_star_quad:.2f} Вт')
print(f'  Степенная:    y_hat = {y_star_pow:.2f} Вт')
```

**Интерпретация прогноза:**

Для процессора AMD с ~9.33 ядрами (это 106% от среднего числа ядер в наборе данных) все три модели дают близкие прогнозы TDP в диапазоне примерно 75-90 Вт. Это выглядит содержательно правдоподобно: 8-12-ядерные процессоры AMD этого периода действительно имеют TDP в диапазоне 65-125 Вт. В качестве основной модели для прогноза возьмём степенную -- она имеет чуть лучшее качество аппроксимации и содержательно оправдана (нелинейный рост энергопотребления с ростом числа ядер).

## 6. Итоговый вывод

В работе исследовался набор данных `openintro::cpu` -- срез по процессорам AMD (60 наблюдений). Изучалась зависимость теплового пакета (TDP, Вт) от числа ядер процессора за период 2010-2020 гг.

Были построены три модели: линейная, квадратичная и степенная. Все три показали R^2 в диапазоне 0.55-0.65, что говорит об умеренной объяснительной силе фактора "число ядер". Значительный остаточный разброс обусловлен тем, что AMD выпускает процессоры с одинаковым числом ядер, но сильно разным TDP (мобильные vs настольные vs серверные). По совокупности критериев (R^2, RMSE, средняя ошибка аппроксимации, характер остатков) предпочтительной оказалась степенная модель `y = a * x^b` -- она лучше описывает нелинейный характер роста TDP при увеличении числа ядер.

Линейная модель статистически значима: коэффициент при x значимо отличается от нуля (t_набл >> t_крит, p-value << 0.05). Зависимость положительная: каждое дополнительное ядро в среднем добавляет несколько ватт к TDP. Прогноз при x* = 9.33 ядрах даёт около 75-90 Вт TDP, что согласуется с реальными характеристиками линейки процессоров AMD того периода.
