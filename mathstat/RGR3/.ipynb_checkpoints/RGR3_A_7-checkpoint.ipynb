# РГР №3 — Регрессионные модели и их интерпретация

**Вариант:** A-7  
**Предметная область:** теплопакет процессора (TDP, Вт) в зависимости от числа ядер — процессоры AMD  
**Источник данных:** набор `openintro::cpu` (CPU's Released between 2010 and 2020)  
Страница: https://vincentarelbundock.github.io/Rdatasets/doc/openintro/cpu.html

**Срез:** фильтр `company == "AMD"`, итого 61 наблюдение

**Переменные:**
- x — число ядер (`cores`)
- y — TDP, Вт (`tdp`)

**Модели для сравнения:**
1. Линейная: y = a + b*x
2. Квадратичная: y = a + b*x + c*x^2
3. Степенная (доп.): y = a * x^b

**Прогнозное значение:** x* = 9.3280 (106% от среднего числа ядер)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

plt.rcParams['figure.figsize'] = (9, 5)
plt.rcParams['axes.grid'] = True
```

## 1. Загрузка и первичный анализ данных

```python
df = pd.read_csv('RGR3_A_7.csv')
x = df['x'].values
y = df['y'].values
n = len(x)

print(f'Количество наблюдений: {n}')
print(f'x (ядра): min={x.min()}, max={x.max()}, mean={x.mean():.4f}')
print(f'y (TDP):  min={y.min()}, max={y.max()}, mean={y.mean():.4f}')
df.head(10)
```

### Диаграмма рассеяния

```python
plt.scatter(x, y, alpha=0.6, edgecolors='k', linewidths=0.5)
plt.xlabel('Число ядер (x)')
plt.ylabel('TDP, Вт (y)')
plt.title('Диаграмма рассеяния: TDP vs число ядер (AMD)')
plt.tight_layout()
plt.show()
```

**Гипотеза о характере связи**

На графике видна положительная связь — с ростом числа ядер TDP в целом увеличивается. При малом числе ядер (2–4) разброс очень большой: процессоры с одинаковым числом ядер могут иметь TDP от 9 до 125 Вт. На больших значениях x (16–64 ядра) данных мало, но TDP заметно выше. Связь скорее нелинейная — рост замедляется или ускоряется неравномерно. Попробуем линейную, квадратичную и степенную модели.

## 2. Построение трёх моделей

### 2.1 Линейная модель: y = a + b*x

```python
x_mean = np.mean(x)
y_mean = np.mean(y)

Sxx = np.sum((x - x_mean)**2)
Sxy = np.sum((x - x_mean) * (y - y_mean))

b1 = Sxy / Sxx
b0 = y_mean - b1 * x_mean

y_hat_lin = b0 + b1 * x

print(f'a (b0) = {b0:.4f}')
print(f'b (b1) = {b1:.4f}')
print(f'Уравнение: y = {b0:.4f} + {b1:.4f} * x')
```

```python
x_line = np.linspace(x.min(), x.max(), 200)

plt.scatter(x, y, alpha=0.6, edgecolors='k', linewidths=0.5, label='данные')
plt.plot(x_line, b0 + b1*x_line, 'r-', lw=2, label=f'линейная: y = {b0:.1f} + {b1:.2f}x')
plt.xlabel('Число ядер')
plt.ylabel('TDP, Вт')
plt.title('Линейная модель')
plt.legend()
plt.tight_layout()
plt.show()
```

### 2.2 Квадратичная модель: y = a + b*x + c*x^2

```python
# решаем через МНК вручную через матрицу дизайна
X_q = np.column_stack([np.ones(n), x, x**2])
coeffs_q = np.linalg.lstsq(X_q, y, rcond=None)[0]
a_q, b_q, c_q = coeffs_q

y_hat_q = a_q + b_q*x + c_q*x**2

print(f'a = {a_q:.4f}')
print(f'b = {b_q:.4f}')
print(f'c = {c_q:.4f}')
print(f'Уравнение: y = {a_q:.4f} + {b_q:.4f}*x + ({c_q:.4f})*x^2')
```

```python
y_q_line = a_q + b_q*x_line + c_q*x_line**2

plt.scatter(x, y, alpha=0.6, edgecolors='k', linewidths=0.5, label='данные')
plt.plot(x_line, y_q_line, 'g-', lw=2, label='квадратичная')
plt.xlabel('Число ядер')
plt.ylabel('TDP, Вт')
plt.title('Квадратичная модель')
plt.legend()
plt.tight_layout()
plt.show()
```

### 2.3 Степенная модель: y = a * x^b

Линеаризация: берём логарифм от обеих частей

`ln(y) = ln(a) + b * ln(x)`

Обозначим Y' = ln(y), X' = ln(x) — тогда это обычная линейная регрессия Y' на X'. Оцениваем МНК, потом возвращаемся к исходному масштабу: a = exp(ln_a).

```python
ln_x = np.log(x)
ln_y = np.log(y)

Sxx_p = np.sum((ln_x - np.mean(ln_x))**2)
Sxy_p = np.sum((ln_x - np.mean(ln_x)) * (ln_y - np.mean(ln_y)))

b_pow = Sxy_p / Sxx_p
ln_a_pow = np.mean(ln_y) - b_pow * np.mean(ln_x)
a_pow = np.exp(ln_a_pow)

y_hat_pow = a_pow * x**b_pow

print(f'a = {a_pow:.4f}')
print(f'b = {b_pow:.4f}')
print(f'Уравнение: y = {a_pow:.4f} * x^{b_pow:.4f}')
```

```python
y_pow_line = a_pow * x_line**b_pow

plt.scatter(x, y, alpha=0.6, edgecolors='k', linewidths=0.5, label='данные')
plt.plot(x_line, y_pow_line, 'm-', lw=2, label=f'степенная: y = {a_pow:.2f} * x^{b_pow:.4f}')
plt.xlabel('Число ядер')
plt.ylabel('TDP, Вт')
plt.title('Степенная модель')
plt.legend()
plt.tight_layout()
plt.show()
```

### Все три модели на одном графике

```python
plt.scatter(x, y, alpha=0.6, edgecolors='k', linewidths=0.5, label='данные', zorder=5)
plt.plot(x_line, b0 + b1*x_line, 'r-', lw=2, label='линейная')
plt.plot(x_line, y_q_line, 'g-', lw=2, label='квадратичная')
plt.plot(x_line, y_pow_line, 'm-', lw=2, label='степенная')
plt.xlabel('Число ядер')
plt.ylabel('TDP, Вт')
plt.title('Сравнение трёх моделей')
plt.legend()
plt.tight_layout()
plt.show()
```

## 3. Сравнение моделей: R^2, RMSE, средняя ошибка аппроксимации

```python
TSS = np.sum((y - y_mean)**2)

def metrics(y_true, y_pred, label):
    res = y_true - y_pred
    RSS = np.sum(res**2)
    R2  = 1 - RSS / TSS
    RMSE = np.sqrt(np.mean(res**2))
    A = 100 / n * np.sum(np.abs(res / y_true))
    print(f'{label}: R2={R2:.4f}, RMSE={RMSE:.4f}, A={A:.2f}%')
    return R2, RMSE, A

R2_lin,  RMSE_lin,  A_lin  = metrics(y, y_hat_lin, 'Линейная    ')
R2_q,    RMSE_q,    A_q    = metrics(y, y_hat_q,   'Квадратичная')
R2_pow,  RMSE_pow,  A_pow  = metrics(y, y_hat_pow, 'Степенная   ')
```

```python
# сводная таблица
comparison = pd.DataFrame({
    'Модель': ['Линейная', 'Квадратичная', 'Степенная'],
    'R^2':    [R2_lin, R2_q, R2_pow],
    'RMSE':   [RMSE_lin, RMSE_q, RMSE_pow],
    'A, %':   [A_lin, A_q, A_pow]
})
comparison = comparison.round(4)
comparison
```

### Графики остатков для трёх моделей

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for ax, y_hat, title in zip(axes,
                             [y_hat_lin, y_hat_q, y_hat_pow],
                             ['Линейная', 'Квадратичная', 'Степенная']):
    res = y - y_hat
    ax.scatter(y_hat, res, alpha=0.6, edgecolors='k', linewidths=0.4)
    ax.axhline(0, color='r', lw=1.5)
    ax.set_xlabel('y_hat')
    ax.set_ylabel('остаток')
    ax.set_title(title)
    ax.grid(True)

plt.tight_layout()
plt.show()
```

**Вывод по сравнению моделей**

По метрикам:
- Линейная: R^2 = 0.7069, RMSE = 33.88, A = 64.58%
- Квадратичная: R^2 = 0.7585, RMSE = 30.75, A = 57.51%
- Степенная: R^2 = 0.7347, RMSE = 32.23, A = 50.24%

Квадратичная модель даёт наилучший R^2 и RMSE. Степенная немного хуже по R^2 и RMSE, но имеет наименьшую среднюю ошибку аппроксимации A = 50.24% — это говорит о том, что она ошибается относительно меньше на малых значениях TDP. Средняя ошибка аппроксимации у всех трёх моделей большая (>50%) — это следствие очень высокой дисперсии TDP внутри групп с одинаковым числом ядер (например, у 4-ядерных процессоров TDP варьируется от 15 до 125 Вт). Это ограничение данных, а не плохой выбор модели.

На графиках остатков структуры нет — остатки разбросаны примерно случайно, что хороший знак. Выбираем **квадратичную модель** как наилучшую по R^2 и RMSE, однако степенная содержательно привлекательна: она отражает типичное для технических систем масштабирование мощности по степенному закону.

## 4. Подробный статистический анализ линейной модели

```python
# остатки и RSS
e = y - y_hat_lin
RSS = np.sum(e**2)
s2 = RSS / (n - 2)  # несмещённая оценка дисперсии
s  = np.sqrt(s2)

print(f'RSS  = {RSS:.2f}')
print(f's^2  = {s2:.4f}')
print(f's    = {s:.4f}')
```

```python
# стандартные ошибки коэффициентов
SE_b1 = s / np.sqrt(Sxx)
SE_b0 = s * np.sqrt(1/n + x_mean**2 / Sxx)

print(f'SE(b1) = {SE_b1:.4f}')
print(f'SE(b0) = {SE_b0:.4f}')
```

```python
# квантиль Стьюдента для 95% ДИ
alpha = 0.05
t_crit = stats.t.ppf(1 - alpha/2, df=n-2)
print(f't_crit(0.975, df={n-2}) = {t_crit:.4f}')

CI_b1 = (b1 - t_crit * SE_b1, b1 + t_crit * SE_b1)
CI_b0 = (b0 - t_crit * SE_b0, b0 + t_crit * SE_b0)

print(f'95% ДИ для b1: ({CI_b1[0]:.4f}, {CI_b1[1]:.4f})')
print(f'95% ДИ для b0: ({CI_b0[0]:.4f}, {CI_b0[1]:.4f})')
```

```python
# проверка гипотезы H0: b1 = 0 vs H1: b1 != 0
t_obs = b1 / SE_b1
p_value = 2 * (1 - stats.t.cdf(abs(t_obs), df=n-2))

print(f't_набл = {t_obs:.4f}')
print(f'|t_набл| = {abs(t_obs):.4f} > t_crit = {t_crit:.4f}: {abs(t_obs) > t_crit}')
print(f'p-value = {p_value:.2e}')

if abs(t_obs) > t_crit:
    print('=> H0 отвергается, коэффициент b1 статистически значим на уровне 0.05')
else:
    print('=> H0 не отвергается')
```

```python
# таблица остатков (первые 10)
resid_df = pd.DataFrame({'x': x, 'y': y, 'y_hat': y_hat_lin.round(4), 'e': e.round(4)})
resid_df.head(10)
```

**Итог по линейной модели**

Коэффициенты: b0 = 46.6372, b1 = 4.7502

95% доверительные интервалы:
- b1: (3.9534, 5.5471) — ноль в интервал не входит
- b0: (35.3708, 57.9036)

t-статистика для b1: t_набл = 11.9285, что значительно больше t_крит = 2.0010. p-value практически равно нулю. Следовательно, **гипотеза H0: b1 = 0 отвергается** — коэффициент при числе ядер статистически значим на уровне 0.05.

Интерпретация: каждое дополнительное ядро в среднем добавляет ~4.75 Вт к TDP процессора AMD.

## 5. Прогноз при x* = 9.3280

```python
x_star = 9.3280

y_lin_pred  = b0 + b1 * x_star
y_q_pred    = a_q + b_q * x_star + c_q * x_star**2
y_pow_pred  = a_pow * x_star**b_pow

print(f'x* = {x_star}')
print(f'Линейная:     y_hat = {y_lin_pred:.4f} Вт')
print(f'Квадратичная: y_hat = {y_q_pred:.4f} Вт')
print(f'Степенная:    y_hat = {y_pow_pred:.4f} Вт')
```

```python
# покажем точку прогноза на графике квадратичной (выбранная модель)
plt.scatter(x, y, alpha=0.6, edgecolors='k', linewidths=0.5, label='данные')
plt.plot(x_line, b0 + b1*x_line, 'r--', lw=1.5, alpha=0.7, label='линейная')
plt.plot(x_line, y_q_line, 'g-', lw=2, label='квадратичная (выбранная)')
plt.plot(x_line, y_pow_line, 'm-', lw=1.5, alpha=0.7, label='степенная')
plt.axvline(x_star, color='orange', linestyle=':', lw=1.5, label=f'x* = {x_star}')
plt.scatter([x_star], [y_q_pred], color='green', s=100, zorder=6, label=f'прогноз = {y_q_pred:.1f} Вт')
plt.xlabel('Число ядер')
plt.ylabel('TDP, Вт')
plt.title('Прогноз при x* = 9.328')
plt.legend(fontsize=8)
plt.tight_layout()
plt.show()
```

## 6. Итоговый вывод

В работе исследовался набор данных `openintro::cpu` — реальные характеристики процессоров AMD, выпущенных с 2010 по 2020 год (61 наблюдение). Изучалась зависимость теплового пакета (TDP, Вт) от числа ядер процессора.

Были построены три модели: линейная, квадратичная и степенная (y = a * x^b). По R^2 и RMSE лучшей оказалась **квадратичная модель** (R^2 = 0.759, RMSE = 30.75), немного превзойдя степенную (R^2 = 0.735) и линейную (R^2 = 0.707). По средней ошибке аппроксимации A лучше всего выглядит степенная (50.24% против 57.51% у квадратичной), но у обеих значения высокие — это объясняется тем, что в данных очень большой разброс TDP внутри каждой группы ядер: например, 4-ядерные процессоры имеют TDP от 15 до 125 Вт. Такая вариация связана с разными сериями (мобильные, серверные, десктопные), которые не разделены в срезе.

Для линейной модели установлено, что коэффициент при числе ядер статистически значим (t_набл = 11.93 >> t_крит = 2.00, p ≈ 0): связь не случайна. Каждое ядро добавляет в среднем ~4.75 Вт TDP.

Прогноз при x* = 9.328 ядер (106% от среднего): по квадратичной модели TDP ≈ 100.4 Вт, по линейной ≈ 90.9 Вт, по степенной ≈ 94.3 Вт. Значения согласуются и попадают в реалистичный диапазон для процессоров с ~10 ядрами (например, Ryzen 9 3900X имеет 12 ядер и TDP 105 Вт).
